# 1. Naloga

```python
import codecs

signs = ['Č', 'Š', 'Ž', 'č', 'š', 'ž']
encodings = ['ibm852', 'iso8859_2', 'cp1250', 'mac_latin2', 'utf_16_le', 'utf_16_be', 'utf_8']

def to_binary(byte_seq):
    return ' '.join(format(byte, '08b') for byte in byte_seq)

for encoding in encodings:
    for sign in signs:
        print(f"Character: {sign}")
        encoded = sign.encode(encoding)
        decimal = [byte for byte in encoded]
        hex_val = [hex(byte) for byte in encoded]
        binary = to_binary(encoded)
        print(f"  Encoding: {encoding}")
        print(f"    Decimal: {decimal}")
        print(f"    Hex: {hex_val}")
        print(f"    Binary: {binary}")
        print()
    print("---------------------")
```

# 2. Naloga

```python
file = open("./kodne točke.txt")
text = file.read()
file.close()

chars = text.split(',')
code_dict = {}
encoded = []
for char in chars:
    encoded.append(chr(int(char)))
    if char not in code_dict.values():
        code_dict[chr(int(char))] = char
                   
encoded = ''.join(value for value in encoded)

outfile = open('./encoded.txt', "w", encoding="utf-8")
outfile.write(encoded)
outfile.close()
```

```python
from collections import OrderedDict
sorted_dict = OrderedDict(sorted(code_dict.items()))
```
